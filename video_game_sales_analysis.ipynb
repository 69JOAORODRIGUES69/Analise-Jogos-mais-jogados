Análise dos Jogos Mais Vendidos no Mundo

Este notebook realiza a análise do dataset `vgsales.csv`, que contém informações sobre os jogos mais vendidos do mundo. Vamos explorar os dados, verificar valores faltantes e identificar os jogos com as maiores vendas globais.

---

Leitura do Arquivo CSV

O dataset será carregado diretamente de um arquivo `.csv`. Utilizaremos o Pandas para manipulação e análise dos dados.

python


```python
# Importar a biblioteca Pandas
import pandas as pd

# Carregar o dataset com o caminho relativo
df = pd.read_csv('vgsales.csv')  # Certifique-se de que o arquivo está na mesma pasta do notebook
df.head()  # Exibir as primeiras linhas

```

```python
# Verificar se há valores faltantes no DataFrame
print(df.isnull().sum())

# Descrever as principais colunas relacionadas às vendas globais
df['Global_Sales'].describe()
```

Top 10 Jogos Mais Vendidos

Vamos agora filtrar os 10 jogos com maiores vendas globais, ordenados de forma decrescente, para identificar os títulos mais vendidos. 

```python
# Ordenar o DataFrame por Global_Sales e exibir os 10 jogos mais Vendidos
top_games = df.sort_values(by='Global_Sales', ascending=False).head(30)
print(top_games[['Name', 'Global_Sales']])
```

Visualização dos Dados

Por fim, vamos gerar um gráfico de barras para visualizar os 10 jogos mais vendidos globalmente. Utilizaremos a biblioteca matplotlib para criar o gráfico. 

```python
# Importar a biblioteca Matplotlib para visualização
import matplotlib.pyplot as plt

# Criar o gráfico de barras com os 10 jogos mais vendidos
top_games.plot(kind='bar', x='Name', y='Global_Sales', title='Top 10 Jogos Mais Vendidos')

# Exibir o gráfico
plt.show()
```

Conclusão
